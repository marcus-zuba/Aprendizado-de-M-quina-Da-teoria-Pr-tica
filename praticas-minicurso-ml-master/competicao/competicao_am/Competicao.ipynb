### b) Estudo dos parâmetros

Agora que sabemos qual o é melhor modelo, SVM, podemos analisar o experimento dele para verificar quais os melhores parâmetros a serem utilizados. Podemos percorrer os resultados de todos os folds utilizados no experimento e pegar a melhor de cada fold para vermos e compararmos os parâmetros.

O estudo dos parâmetros foi realizado utilizando a superclasse ObjetivoOtimização, que foi herdada para 3 diferentes classes: OtimizacaoObjetivoSVMCompeticao, OtimizacaoObjetivoRandomForest e OtimizacaoObjetivoArvoreDecisao, que contemplam os 3 métodos que foram utilizados nesse experimento. 

```python
exp = experimentos[0]
#Pego o primeiro experimento pq ele é o experimento do SVM, que foi escolhido como melhor

#crio um dataframe vazio 
study_all_folds = pd.DataFrame()

#percorro os study de cada fold e pego a trial com o maior valor do macroF1 de cada um 
for i in range(5):
    study_fold = exp.studies_per_fold[i]
    df_study = study_fold.trials_dataframe().sort_values("value",ascending=False, ignore_index=True)
    study_all_folds = study_all_folds.append(df_study[df_study.index==0], ignore_index=True)

#adiciono uma coluna com a identificação dos folds pra ser o index
folds_labels = [1, 2, 3, 4, 5]
study_all_folds['fold']=folds_labels
study_all_folds.set_index('fold', inplace=True)
```

Sabemos que o min_samples_split afetará o valor do parâmetro "C" do modelo SVM, cuja fórmula se dá por: 2^(min_samples_split)b

```python
valores_min_samples = [0.693396, 1.027763, 1.827388, 0.224118, 0.007595]
```

```python
import pandas as pd
from avaliacao_competicao import testar_parametros
```

```python
df = pd.read_csv("../datasets/movies_amostra.csv")

medias=[[],[]]

for min_sample in valores_min_samples:
    
    media_acuracia = 0
    media_macroF1 = 0
    for i in range(10):

        df_treino = df.sample(frac=0.8, random_state=i)
        df_teste = df.drop(df_treino.index)

        resultado = testar_parametros(df_treino, df_teste, min_sample)

        media_acuracia += resultado.acuracia
        media_macroF1 += resultado.macro_f1

    media_acuracia/=10
    media_macroF1/=10
        
    medias[0].append(media_acuracia)
    medias[1].append(media_macroF1)
```

```python
i = 0
for min_sample in valores_min_samples:

    print(f'\nMIN SAMPLE={min_sample}\n')
        
    print(f'Média acurácia = {medias[0][i]}')
    print(f'Média macroF1 = {medias[1][i]}')
    
    i+=1
```

Definido o melhor modelo, as predições foram realizadas novamente apenas com o método que obteve o maior desempenho nos testes anteriores. Dessa forma, é possível analisar os valores de min_samples e determinar o melhor valor desse parâmetro ao se observar as médias da acurácia e do macro_f1 dos resultados retornados pelas predições.
